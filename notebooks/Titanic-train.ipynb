```python
df['Age_filled'] = df['Age'].fillna(0) 
# создание нового столбца 'Age_filled', где пропущенные значения заменены на 0
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
# заполнение пропущенных значений в столбце 'Age' медианным значением возраста 
df
```

```python
most_freq = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(most_freq) 
# заполнение пропущенных значений в столбце 'Embarked' наиболее частым значением
df
```

```python
df['Embarked'].value_counts() # количество пассажиров в каждом порту посадки
```

```python
df.groupby('Pclass')['Fare'].mean().round(2)
# средний тариф пассажиров в каждом классе 
```

```python
df.groupby('Sex')['Age'].mean().round(2)
# средний пол пассажиров в каждом классе
```

```python
df.groupby(['Sex', 'Pclass'])['Age'].mean().round(2)
# средний возраст пассажиров в зависимости от пола и класса
```

```python
df.groupby('Pclass')['Fare'].agg(['min', 'max', 'mean', 'median']).round(2) 
# минимальный, максимальный, средний и медианный тариф пассажиров в каждом классе
```

```python
df.groupby('Pclass')['Sex'].value_counts() 
# количество пассажиров каждого пола в каждом классе
```
