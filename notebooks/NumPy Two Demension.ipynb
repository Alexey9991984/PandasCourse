```python
import numpy as np 
```

```python
first = np.array([[1, 2, 3], [4, 5, 6,]]) # 2D array
first
```

```python
first.shape 
# размерность массива (2 строки, 3 столбца)
```

```python
first.ndim
# количество осей (измерений) массива
```

```python
second = np.array([[7, 8, 9], [10, 11, 12]]) # 2D array
second
```

```python
second.shape 
# размерность массива (2 строки, 3 столбца)
```

```python
first + second
# поэлементное сложение двух массивов
```

```python
first + 1 # прибавление скаляра ко всем элементам массива 
```

```python
first - 2 # вычитание скаляра из всех элементов массива
```

```python
first * 2 # умножение всех элементов массива на скаляр
```

```python
first / 3 # деление всех элементов массива на скаляр
```

```python
first # получение всех элементов массива
```

```python
first[1, 1]  # получение элемента из второй строки и второго столбца
```

```python
np.concatenate((first, second)) 
# объединение двух массивов по строкам (по умолчанию вертикально axis=0)
```

```python
np.concatenate((first, second), axis=1) 
# объединение двух массивов по столбцам (горизонтально axis=1)
```

```python
np.concatenate((np.array([1,2,3,4]), np.array([5,6,7,8]))) 
# объединение одномерных массивов по умолчанию вертикально (axis=0)
# одномерные массивы не имеют столбцов, поэтому axis=1 здесь не применимо
```
