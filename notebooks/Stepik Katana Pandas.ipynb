## DataFrame

```python
information = {"Имя": ["Алиса", "Боб", "Катя"],
           "Возраст": [25, 30, 22],
             "Город": ["Киото", "Эдо", "Осака"]
}
df = pd.DataFrame(information, index=[i for i in range(1, 4)]) 
index=[i for i in range(1, 4)] # задает индексы строк с 1 по 3. 
# Позволяет из словаря сделать таблицу состоящую из нескольких столбцов.
df
```

```python
df.head(10) # Показывает первые 10 строк таблицы.
```

```python
df.tail(3) # Показывает последние 3 строки таблицы.
```

```python
df.shape # Показывает размер таблицы (строки, столбцы).
```

```python
df.info() 
# Показывает информацию о таблице: количество строк, столбцов, типы данных.
```

```python
df.describe() # Показывает статистические данные по числовым столбцам таблицы.
```
