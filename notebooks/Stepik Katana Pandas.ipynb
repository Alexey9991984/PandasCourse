```python
df = pd.DataFrame({
    'animal': ['cat', 'dog', 'cat', 'dog'],
    'age': [2, 7, 1, 5]
})
df
```

```python
mask = df['animal'] == 'cat'  
# Создает булеву маску для строк, где значение в столбце 'animal' равно 'cat'.
df[mask] # Применяет маску к DataFrame, возвращая 
# только строки, где условие маски истинно. 
```

```python
df[df['animal'] == 'cat']
# То же самое, что и выше, но без создания отдельной переменной mask.
```

```python
df[(df['animal'] == 'cat') & (df['age'] < 3)]
# Фильтрует строки, где 'animal' равно 'cat' и 'age' меньше 3.
```

```python
df = pd.DataFrame({'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
                   'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
                   'name': ['Murzik', 'Pushok', 'Kaa', 'Bobik', 'Strelka', 'Vaska', 'Kaa2', 'Murka', 'Graf', 'Muhtar'],
                   'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
                   'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']},
                    index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
filter_names = ["animal", "age"] # Названия столбцов для фильтрации.
filter_values = ["cat", 3] # Значения для фильтрации по столбцам.
df 
```

```python
df[(df[filter_names[0]] == filter_values[0]) & (df[filter_names[1]] < filter_values[1])]
# Фильтрует строки таблицы df, где значение в столбце, 
# название которого хранится в filter_names[0], равно filter_values[0], 
# и значение в столбце, название которого хранится в filter_names[1],
# меньше filter_values[1].  
```
