```python
df2 = pd.read_sql("""SELECT
    title,
    author,
    amount,
    (SELECT MAX(amount) FROM book) - amount AS Заказ
FROM book
WHERE amount < (SELECT MAX(amount) FROM book)""", engine)       
df2 # Вывести название книги, автора, количество экземпляров в наличии и 
# сколько нужно заказать до максимального количества экземпляров
# для всех книг, у которых количество экземпляров меньше максимального
#  количества экземпляров среди всех книг
```

```python
df1 = pd.read_sql("""SELECT author, title, price
FROM book
WHERE price <= (
    SELECT AVG(price)
    FROM book)
ORDER BY price DESC""", engine)
df1 # Вывести автора, название и цену всех книг, цена которых меньше
#  или равна средней цене всех книг, отсортировав результат по убыванию цены    
```
