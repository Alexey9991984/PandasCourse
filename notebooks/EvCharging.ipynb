```python
fixes = {
    "warzawa": "Warszawa",
    "wraszawa": "Warszawa",
    "warsaw": "Warszawa",
    "gdansk": "Gdańsk",
    "krakow": "Kraków",
    "lodz": "Łódź",
    "Crakow": "Kraków",
    "katowicee": "Katowice",
    "poznan": "Poznań",
    "szczecin": "Szczecin",
    "wroclaw": "Wrocław",
    "Katovice": "Katowice",
    "Warsaw": "Warszawa"        # добавь сюда свои ошибки
}

df["city"] = (
    df["city"]
        .str.strip()
        .str.title()
        .replace(fixes)
)
```
