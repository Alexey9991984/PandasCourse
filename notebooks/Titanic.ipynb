```python
mdf = pd.DataFrame(index=df.index)  # Создает новый пустой DataFrame с таким же индексом, как у df
mdf['PassengerID'] = df['PassengerId']  # Копирует столбец 'PassengerId' из df в mdf
mdf['evenID'] = mdf['PassengerID'].apply(lambda x: x % 2 == 0) 
# Создает новый столбец 'evenID' в mdf, где значение True, если 'PassengerID' четный
mdf
```

```python
pd.merge(df, mdf, how='inner', left_index=True, right_index=True)
# Выполняет внутреннее объединение таблиц df и mdf по общему индексу)
```

```python
tmp_df = df.copy(deep=True) # Создает глубокую копию DataFrame df и сохраняет ее в tmp_df
```

```python
tmp_df['Pclass'] = 1 # Задает значение 1 для всех записей в столбце 'Pclass' в tmp_df
```

```python
tmp_df['Pclass'].value_counts() # Считает количество уникальных значений в столбце 'Pclass' в tmp_df
```

```python
df['Pclass'].value_counts() 
# Считает количество уникальных значений в столбце 'Pclass' в df
```

```python
tmp_df['IsAdult'] = tmp_df['Age'] >= 18 # Создает новый столбец 'IsAdult' в tmp_df,
# где значение True, если возраст пассажира 18 лет и старше
tmp_df.head(10)
```

```python
tmp_df['IsAdultSurvived'] = tmp_df['IsAdult'] # Создает новый столбец 'IsAdultSurvived' в tmp_df,
tmp_df.loc[tmp_df['Survived'] == 0, 'IsAdultSurvived'] = False 
# устанавливает значение False в столбце 'isSurvived' для всех пассажиров, которые не выжили
```

```python
tmp_df.head(20) 
# Выводит первые 20 строк таблицы tmp_df
```

```python
tmp_df.rename(columns={'IsAdultSurvived': 'AdultSurvived'}, inplace=True)
# Переименовывает столбец 'IsAdultSurvived' в 'AdultSurvived' в tmp_df
```
