```python
import numpy as np
```

```python
np.arange(10)  
# создает одномерный массив с числами от 0 до 9 
```

```python
np.arange(15, 25) 
# создает одномерный массив с числами от 15 до 24
```

```python
np.arange(20, 30, 3)
# создает одномерный массив с числами от 20 до 29 с шагом 3
```

```python
np.arange(start=-15, stop=5, step=4) 
# создает одномерный массив с числами от -15 до 4 с шагом 4
```

```python
np.arange(20).reshape([2, 10]) 
# преобразует одномерный массив из 20 элементов в двумерный массив 
# размером 2 строки на 10 столбцов
```

```python
np.arange(20).reshape([2, 2, 5]) 
# преобразует одномерный массив из 20 элементов в трехмерный массив 
# размером 2 блока на 2 строки и 5 столбцов
```

```python
np.arange(20).reshape([4,5]) 
# преобразует одномерный массив из 20 элементов в двумерный массив
# размером 4 строки на 5 столбцов
```

```python
np.random.random((10)).reshape([2,5])
# создает одномерный массив из 10 случайных чисел от 0 до 1 и
# преобразует его в двумерный массив размером 2 строки на 5 столбцов
```

```python
np.random.seed(10) 
# устанавливает начальное значение генератора
# случайных чисел для воспроизводимости
np.random.random((10)).reshape([2,5]) 
# создает одномерный массив из 10 случайных чисел от 0 до 1 и
# преобразует его в двумерный массив размером 2 строки на 5 столбцов    
```

```python
np.random.seed(10)
np.random.randint(50, 80, (4, 3, 3)).flatten()
# создает трехмерный массив размером 4 блока на 3 строки и 3 столбца
# с случайными целыми числами от 50 до 79
# и преобразует его в одномерный массив при помощи flatten
```
