```python
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
values = [0, 1, 2, 3, 4, 5, 6, 7] # создал список
squares = [value ** 2 for value in values] # для возведения значения во 2 степень
squares # возвел во 2 степень все цифры из списка values     
```

```python
plt.plot(values, squares) # первая созданная диаграмма по значениям
plt.show()  
```

```python
plt.scatter(values, squares) # еще один вид диаграммы
plt.show()
```

```python
values = list(range(100)) # создание списка с диапозона от 0 до 100
pows = [pow(value, 3)for value in values] # возведение в 3 степень
plt.scatter(values, pows) # создание диаграммы метод scatter
plt.show() # вызов диаграммы
```

```python
plt.scatter(values, pows, c=pows, cmap=plt.cm.Blues, edgecolor='none', s=10)
plt.show()
```
