```python
import pandas as pd
import numpy as np
```

```python
forum_users = {
    'User_ID': np.array([1, 2, 3, 4, 5]),
    'Username': ['Bogdan_s', 'Jane_Smith', 'Alex123', 'Bob56', 'Mark_Wilson'],
    'Age': [18, 35, 25, 38, 33],
    'Joined date': pd.to_datetime(['2032-01-01', '2032-02-10', '2032-01-19', '2032-01-25', '2032-03-15']),
    'Total Posts': [150, 230, 80, 420, 310],
    'Reputation': [500, 720, 200, 940, 500]
} 

df = pd.DataFrame(forum_users)    
df 
```

## Filtring 

```python
df.loc[3] # вывод строки с индексом 3 (Series)
```

```python
df.loc[2:3] # вывод строк с индексами от 2 до 3 включительно (DataFrame)
```

```python
df.loc[2:3, ['Username', 'Joined date', 'Reputation']]
 # вывод строк с индексами от 2 до 3 включительно и столбцов 
 # Username, Joined date и Reputation
```

```python
df.iloc[1] # вывод строки с индексом 1 (Series)
```

```python
df.iloc[2:4] # вывод строк с позициями от 2 до 3 включительно (DataFrame)
```

```python
df.iloc[2:4, 1:4]
```

```python
df.iloc[:, 1:4] # вывод всех строк и столбцов с позициями от 1 до 3 включительно
```

```python
df[df['Age'] >= 25] # пользователи старше 25 лет
```

```python
df['Age'] >= 25 # булев массив для фильтрации (Series)
```

```python
df[df['Total Posts'] > 300] # пользователи с количеством постов больше 300
```

```python
df[df['Reputation'] < 500] # пользователи с репутацией меньше 500
```

```python
df[(df['Total Posts'] >= 300) & (df['Age'] >= 25)] # пользователи старше 25 лет с количеством постов больше или равно 300
```

```python
(df['Total Posts'] >= 300) & (df['Age'] >= 25) # пользователи старше 25 лет с количеством постов больше или равно 300
# (Series булев массив для фильтрации)
```

```python
df[(df['Total Posts'] >= 400) | (df['Age'] <= 25)] # пользователи с количеством постов больше или равно 300 или младше 25 лет    
```

```python
df['Reputation'].isin([200, 500]) # пользователи с репутацией 200 или 500 
# Serial
```

```python
df[df['Reputation'].isin([200, 500])] # пользователи с репутацией 200 или 500
# df filter
```

```python
df[df['Reputation'].isin(range(200, 500))] # пользователи с репутацией 200 или 500
```

```python
df[df['Total Posts'].isin(range(200, 350))] # пользователи с количеством постов от 200 до 349
```

```python
date_range = pd.date_range(start='2032-01-15', end='2032-02-15')  
date_range # диапазон дат переменной date_range
```

```python
df[df['Joined date'].isin(date_range)]  # пользователи, присоединившиеся в период с 15 января по 15 февраля 2032 года
```

## Sorting 

```python
df.sort_values(by='Age') # сортировка по возрастанию возраста
```

```python
df.sort_values(by='Joined date', ascending=False)  # сортировка по убыванию даты присоединения
```

```python
df[['User_ID', 'Total Posts', 'Reputation']]
```

```python
df[['User_ID', 'Total Posts', 'Reputation']].sort_values(by=2, axis=1, ascending=False,) 
# сортировка столбцов по убыванию количества постов и возрастанию репутации 
# по оси столбцов (axis=1)
```

```python
df.sort_values(by='Total Posts') # сортировка по возрастанию количества постов 
```
