```python
import numpy as np
```

```python
np.zeros((3, 4)) 
# создает массив из нулей размером 3x4
```

```python
np.zeros((2, 4), dtype=int) 
# создает массив из нулей размером 2x4 с типом данных int
```

```python
np.ones((5,3), dtype=int)  
# создает массив из единиц размером 5x3 с типом данных int 
```

```python
np.random.seed(0) 
# устанавливает начальное значение генератора случайных 
# чисел для воспроизводимости результатов 
# этот метод нужно вызывать в той же ячейке где 
# генерируются случайные числа
np.random.random((2,4)) 
# создает массив размером 2x4 с случайными числами от 0 до 1 
```
