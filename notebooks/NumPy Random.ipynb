```python
import numpy as np
```

```python
np.zeros((3, 4)) 
# создает массив из нулей размером 3x4
```

```python
np.zeros((2, 4), dtype=int) 
# создает массив из нулей размером 2x4 с типом данных int
```

```python
np.ones((5, 3), dtype=int)  
# создает массив из единиц размером 5x3 с типом данных int 
```

```python
np.random.seed(0)
# устанавливает начальное значение генератора случайных 
# чисел для воспроизводимости результатов 
# этот метод нужно вызывать в той же ячейке где 
# генерируются случайные числа
np.random.random((2, 4)) 
# создает массив размером 2x4 с случайными числами от 0 до 1 
```

```python
np.random.seed(1) 
# устанавливает начальное значение генератора случайных чисел
np.random.randint(10, 20, size=(2, 4))
 # создает массив размером 2x4 с случайными целыми числами от 10 до 19
```

```python
np.random.seed(2) 
# устанавливает начальное значение генератора случайных чисел
np.random.uniform(10.0, 50.0, size=(3, 5)) 
# создает массив размером 3x5 с случайными числами с 
# плавающей точкой от 10.0 до 50.0
```

```python
np.random.seed(3) 
# устанавливает начальное значение генератора случайных чисел
np.random.choice(np.array([1,2,3]), size=10) 
# создает одномерный массив из 10 случайных элементов, выбранных из 
# переданного массива [1,2,3] с равной вероятностью   
```

```python
np.random.seed(1)
# устанавливает начальное значение генератора случайных чисел
np.random.choice(np.array(['a','b','c','d']), size=15) 
# создает одномерный массив из 15 случайных элементов, выбранных из 
# переданного массива ['a','b','c','d'] с равной вероятностью
```
