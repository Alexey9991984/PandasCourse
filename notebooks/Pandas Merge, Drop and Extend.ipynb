```python
import pandas as pd
import numpy as np
```

```python
forum_users = {
    'User_ID': np.array([1, 2, 3, 4, 5]),
    'Username': ['Bogdan_s', 'Jane_Smith', 'Alex123', 'Bob56', 'Mark_Wilson'],
    'Age': [18, 35, 25, 38, 33],
    'Joined date': pd.to_datetime(['2032-01-01', '2032-02-10', '2032-01-19', '2032-01-25', '2032-03-15']),
    'Total Posts': [150, 230, 80, 420, 310],
    'Reputation': [500, 720, 200, 940, 500]
} 

df = pd.DataFrame(forum_users)    
df 
```

```python
df['Language'] = 'English' # Добавление столбца с одинаковыми значениями
df
```

```python
df['Active'] = [True, True, False, True, False] 
df # Добавление столбца с разными значениями
```

```python
total_comments = {
    'Total Comments': [320, 450, 120, 600, 410]
    }

df2 = pd.DataFrame(total_comments)
df2  # Создание второго датафрейма
```

```python
pd.concat([df, df2]) # Объединение двух датафреймов по строкам (по умолчанию) 
# ось = 0
```

```python
df = pd.concat([df, df2], axis=1) # присоединение двух датафреймов по столбцам (ось = 1)
df # Объединение двух датафреймов по столбцам (ось = 1)
```

```python
# пример df.drop(['Language']) # Удаление столбца без сохранения изменений в исходном датафрейме
# ось = 1 по умолчанию, так как поля по оси 1 такого нет, то удаление не происходит
# Исходный датафрейм остался без изменений
```

```python
df.drop(['Language'], axis=1) # Удаление столбца
```

```python
df # Просмотр датафрейма после удаления столбца где видно что столбец остался
```

```python
df.drop(['Language'], axis=1, inplace=True) # Удаление столбца с сохранением изменений в исходном датафрейме
```

```python
df # Просмотр датафрейма после удаления столбца где видно что столбец удален
```

```python
df.drop([2, 3], inplace=True) # Удаление строк 
```

```python
df # Просмотр датафрейма после удаления столбца где видно что столбец удален
```
