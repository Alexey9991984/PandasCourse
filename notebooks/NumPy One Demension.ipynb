```python
import numpy as np
```

```python
np # использование библиотеки NumPy для работы с массивами и числовыми операциями
```

```python
nums_array = np.array([1,2,3,4]) 
# cоздание одномерного массива NumPy из списка Python   
nums_array
```

```python
type(nums_array)
 # проверка типа созданного объекта
```

```python
nums_array[0] 
# доступ к первому элементу массива
```

```python
nums_array[-1]  
# доступ к последнему элементу массива
```

```python
nums_array.shape 
# получение формы массива (количество элементов в каждом измерении)
```

```python
nums_array.ndim  # получение количества измерений массива (одномерный массив имеет 1 измерение)
```

```python
nums_array.size 
# получение общего количества элементов в массиве
```

```python
nums_array.dtype 
# получение типа данных элементов массива
```

```python
nums_array.itemsize 
# получение размера в байтах каждого элемента массива
```

```python
str_array = np.array([True, 10, 'abc', 5.5]) 
str_array 
# создание массива с разными типами данных и его отображение
```

```python
str_array.dtype.name 
# получение типа данных элементов массива в виде строки
```

```python
str_array.itemsize 
# получение размера в байтах каждого элемента массива   
```

```python
obj_array = np.array([{'city': 'London'}, {'city', 'Boston'}])
obj_array
# создание массива объектов и его отображение 
```

```python
obj_array.dtype.name 
# получение типа данных элементов массива в виде строки
```

```python
obj_array.itemsize 
# получение размера в байтах каждого элемента массива  
```
